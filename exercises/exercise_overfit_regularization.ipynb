```python
import pandas as pd 
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
tips= sns.load_dataset("tips")
tips
```

```python
tips.info()
```

```python
y= tips["tip"]
x= pd.get_dummies(tips.drop("tip", axis= "columns"), columns= ["sex", "smoker", "day", "time"])
```

```python
x
```

```python
y
```

```python
X_train, X_test, y_train, y_test= train_test_split(x, y, test_size= 0.3, random_state= 42)
```

```python
print(f"{X_train.shape= }")
print(f"{y_train.shape= }")
print(f"{X_test.shape= }")
print(f"{y_test.shape= }")
```

```python
scaler= MinMaxScaler()

scaler.fit(X_train)
```

```python
scaled_X_train= scaler.transform(X_train)
scaled_X_test= scaler.transform(X_test)
```

```python
print(f"{scaled_X_train.shape= }")
print(f"{scaled_X_train.min()= }")
print(f"{scaled_X_train.max()= }")

print(f"{scaled_X_test.shape= }")
print(f"{scaled_X_test.min()= }")
print(f"{scaled_X_test.max()= }")
```

```python
model= LinearRegression()
model.fit(scaled_X_train, y_train)
```

```python
print(f"{model.coef_= }")
print(f"{model.intercept_= }")
```

```python
test_sample_features= scaled_X_test[0]
test_sample_labels= y_test.values[0]

print(f"{test_sample_labels= }")
print(f"{test_sample_features= }")
```

```python
model.predict(test_sample_features.reshape(1, -1))
```

```python
y_pred= model.predict(scaled_X_test)
y_pred
```

```python
y_test.to_numpy()
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
ramse= np.sqrt(mean_squared_error(y_test, y_pred))
```
